```python
import pandas as pd
```

```python
train = pd.read_csv('Data_Train.csv')
```

```python
train.head()
```

```python
train.info()
```

### The New_Price column is removed as it has only very few values


```python
train.drop(columns=['New_Price'],inplace=True)
```

```python
train.head()
```

```python
train.sort_values(['Kilometers_Driven'],ascending=False,inplace=True)
train
```

### Here the highest value of the Kilometers_Driven is removed as it could become a possible outlier

```python
train = train.drop([2328])
```

```python
import seaborn as sns
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

### The string values that are important for prediction are converted into integers by labeling it

```python
train['Location'] = le.fit_transform(train['Location'])
train['Owner_Type'] = le.fit_transform(train['Owner_Type'])
train['Transmission'] = le.fit_transform(train['Transmission'])
train['Fuel_Type'] = le.fit_transform(train['Fuel_Type'])
```

### The columns below mentioned are numbers in string which are processed into integers

```python
train['Engine'].fillna('0000 cc',inplace=True)
train['Mileage'].fillna('0000 kmpl',inplace=True)
train['Power'].fillna('000 bhp',inplace=True)
train['Seats'].fillna(int(train['Seats'].mean()),inplace=True)
```

```python
a=list()
for i in train['Mileage']:
    a.append(float(i[:4]))
train["Mileage"]=a
   
```

```python
k1=list()
for i in train['Mileage']:
    if i==0.0:
        i=train['Mileage'].mean()
    k1.append(i)
train['Mileage']=k1
```

```python
b = list()
for i in train['Engine']:
    if(i=='72 CC'):
        i='72  '
    b.append(int(i[:4]))
train["Engine"]=b
```

```python
k2=list()
for i in train['Engine']:
    if i==0.0:
        i=train['Engine'].mean()
    k2.append(i)
train['Engine']=k2
```

```python
c = list()
for i in train['Power']:
    if i=='null bhp':
        i='0000'
    c.append(i[:5])
d = list()
for i in c:
    if 'bh' in i:
        i=i[:2]
    if 'b' in i:
        i=i[:3]
    d.append(float(i))
train['Power']=d
```

```python
k3=list()
for i in train['Power']:
    if i==0.0:
        i=train['Power'].mean()
    k3.append(i)
train['Power']=k3
```

```python
y=train['Price']
train.drop(columns=['Price'],inplace=True)
train1 = train.drop(columns=['Name'])
```
